```python
one_hot_encode = pd.get_dummies(data, columns = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Credit_History', 'Property_Area'])
```

```python
one_hot_encode
```

```python
one_hot_encode = one_hot_encode.reset_index(drop = True)
```

```python
data_preprocessed = pd.DataFrame(one_hot_encode)
```

```python
data_preprocessed
```

```python
preprocessed_train = data_preprocessed[data_preprocessed['Loan_Status'].notna()]
preprocessed_test = data_preprocessed[data_preprocessed['Loan_Status'].isna()]
```

```python
data_preprocessed.info()
```

```python
x = preprocessed_train.drop(columns = ['Loan_ID', 'Loan_Status'])
y = preprocessed_train['Loan_Status']
```

### Train_test_split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 40)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### LogisticRegression

```python
model = LogisticRegression(max_iter = 500)
```

```python
model.fit(x_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

### Train Prediction

```python
y_pred = model.predict(x_train)
```

```python
print('accuracy_score:',accuracy_score(y_train, y_pred))
print('f1_score:', f1_score(y_train, y_pred, pos_label='Y'))
print('confusion_matrix:\n', confusion_matrix(y_train, y_pred))
```

### Test Prediction

```python
y_pred_test = model.predict(x_test)
```

```python
print('accuracy_score:',accuracy_score(y_test, y_pred_test))
print('f1_score:', f1_score(y_test, y_pred_test, pos_label='Y'))
print('confusion_matrix:\n', confusion_matrix(y_test, y_pred_test))
```

```python
preprocessed_x_test = preprocessed_test.drop(columns=['Loan_ID','Loan_Status'])
y_prediction = model.predict(preprocessed_x_test)
```

```python
sample_submission['Loan_Status'] = y_prediction
sample_submission.to_csv('submission_loan_prediction.csv',index=False)
```
